you are driving too fast, and a police officer stops you. write a function to return one of the three possible results:'no ticket','small ticket' and 'big ticket'. if your speed is 60 or less, the result is 'no ticket'.if speed is bet. 61 and 80 inclusive, the result is 'small ticket'.....

```python
def caught_speeding(speed,is_birthday):
    if is_birthday:
        speeding = speed - 5
    else:
        speeding = speed
        
    if speeding > 80:
        return 'Big Ticket'
    elif speeding > 60:
        return 'Small Ticket'
    else:
        return 'No ticket'
        
        
        
```

```python
caught_speeding(81,True)
```

```python
caught_speeding(81,False)
```
